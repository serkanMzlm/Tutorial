```python
import numpy as np
import pandas as pd

# Modeller
from sklearn.ensemble import RandomForestRegressor 
from sklearn.ensemble import RandomForestClassifier 
from sklearn.linear_model import Ridge
from sklearn.svm import LinearSVC

from sklearn.model_selection import train_test_split  # Veri setinin bölmeye yarar

from sklearn.datasets import fetch_california_housing  # Sklearn sunduğu hazor data

housing = fetch_california_housing()
housing
```

```python
housing_df = pd.DataFrame(housing["data"], columns=housing["feature_names"]) # Data bulunan veriler yapay zekaya uygun hale getirilir.
housing_df.head()
```

```python
housing_df["target"] = housing["target"]
housing_df.head()
```

---

`Ridge` ve `RandomForestRegressor` modelleri denendi ve bu örnek için `RandomForestRegressor` daha iyi sonuç çıkardı. Model seçmek için sklearn sayfasında bulunan map izlenebilir.

---

```python
np.random.seed(42)

X = housing_df.drop("target", axis=1)
y = housing_df["target"]

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)

model1 = Ridge()
model2 = RandomForestRegressor()


model1.fit(X_train, y_train)
model2.fit(X_train, y_train)
ridge_model = model1.score(X_test, y_test)
random_model = model2.score(X_test, y_test)

print(f"Ridge Model Coef: {ridge_model*100:.2f}")
print(f"RFR Model Coef: {random_model*100:.2f}")
```

---

`LinearSVC` ve `RandomForestClassifier` modelleri denendi ve bu örnek için `RandomForestRegressor` daha iyi sonuç çıkardı. Model seçmek için sklearn sayfasında bulunan map izlenebilir.

---

```python
heart_disease = pd.read_csv("data/heart-disease.csv")
heart_disease.isna().sum(),"-"*30, len(heart_disease)
```

```python
heart_disease.head()
```

```python
heart_disease["target"].value_counts()
```

```python
np.random.seed(42)

X = heart_disease.drop("target", axis=1)
y = heart_disease["target"]

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)


clf = LinearSVC(max_iter=10000)
clf.fit(X_train, y_train)
linear_svc = clf.score(X_test, y_test)

clf = RandomForestClassifier(n_estimators=100)
clf.fit(X_train, y_train)
rfc = clf.score(X_test, y_test)

print(f"LinearSVC Model Coef: {linear_svc*100:.2f}")
print(f"RFC Model Coef: {rfc*100:.2f}")
```
